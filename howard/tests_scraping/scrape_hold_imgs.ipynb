```python
import os
import requests
```

```python
def download_img(image_url, image_path):
    img_data = requests.get(image_url).content
    with open(image_path, 'wb') as handler:
        handler.write(img_data)
    return

def get_hold_names():
    holds1 = ['h'+str(i)+'.png' for i in list(range(1,201))]
    holds2 = []
    holds2_str = [str(i) for i in list(range(1,81))]
    for h in holds2_str:
        if len(h)==1:
            h='0'+h
        holds2.append('hw'+str(h)+'.png')
    holds = holds1 + holds2
    return holds
```

```python
parent_url = 'https://www.moonboard.com/content/images/holds/'
img_dir = '/Users/aaronwu/Documents/misc/MoonBoard/data/assets/hold_imgs/'
```

```python
hold_names = get_hold_names()
```

```python
for hold_name in hold_names:
    print(hold_name)
    image_url = parent_url+hold_name
    image_path = img_dir+hold_name
    if not os.path.exists(img_dir+hold_name):
        download_img(image_url, image_path)
```
